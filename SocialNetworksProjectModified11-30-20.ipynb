```python
import tweepy
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import numpy as np
import matplotlib.pyplot as plt

consumer_key = "D55jTaBZl3rAmYMg0d5vh4hh1"
consumer_secret = "mObfw4wZ6VscSQ9SHcuXVRHwySfTKV12wLTb9zHAeBUAvOZoKH"
access_token = "1330915157631909889-TVOgTZ6gZjpuTObK9DLuihKUnp16Tk"
access_token_secret = "kGR3LVMdAEnBQHqfSXPFtv4OEel1uapN2ODzPzzhuR7Ah" 

#####################################
# Authentication process 

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

# test authentication
try:
    api.verify_credentials()
    print("Authentication OK")
except:
    print("Error during authentication")

#####################################
# Get tweets from home timeline

ids = []
usernames = []
tweets = []
idcount = 0

# get tweets

# first half of tweets
# political personas
twitobj = api.user_timeline(screen_name = 'hasanthehun', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('hasanthehun')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'IlhanMN', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('IlhanMN')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'CNN', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('CNN')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'BernieSanders', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('BernieSanders')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'JoeBiden', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('JoeBiden')
    ids.append(idcount)
    idcount += 1
# basketball players
twitobj = api.user_timeline(screen_name = 'KingJames', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('KingJames')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'isaiahthomas', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('isaiahthomas')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'raf_tyler', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('raf_tyler')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'DwightHoward', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('DwightHoward')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Giannis_An34', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Giannis_An34')
    ids.append(idcount)
    idcount += 1
# Gamers
twitobj = api.user_timeline(screen_name = 'shroud', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('shroud')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'drdisrespect', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('drdisrespect')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'timthetatman', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('timthetatman')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'summit1g', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('summit1g')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'pewdiepie', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('pewdiepie')
    ids.append(idcount)
    idcount += 1
# food people
twitobj = api.user_timeline(screen_name = 'GordonRamsay', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('GordonRamsay')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'seanseaevans', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('seanseaevans')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'CakeBossBuddy', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('CakeBossBuddy')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'mikexingchen', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('mikexingchen')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'tasty', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('tasty')
    ids.append(idcount)
    idcount += 1
# random
twitobj = api.user_timeline(screen_name = 'MakeAnything3D', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('MakeAnything3D')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'uppermostmusic', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('uppermostmusic')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'FallGuysGame', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('FallGuysGame')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'GMHikaru', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('GMHikaru')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'tonyhawk', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('tonyhawk')
    ids.append(idcount)
    idcount += 1
    
# Second half of tweets
twitobj = api.user_timeline(screen_name = 'BarackObama', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('BarackObama')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Bill Clinton', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Bill Clinton')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Merone', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Merone')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Kamala Harris', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Kamala Harris')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Ms. Turnage Young', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Ms. Turnage Young')
    ids.append(idcount)
    idcount += 1

#set 2 internet personality- 5 persons
twitobj = api.user_timeline(screen_name = 'xQc', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('xQc')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'NRG Ronaldo', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('NRG Ronaldo')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'hungrybox', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('hungrybox')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Alexandria Ocasio-Cortez', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Alexandria Ocasio-Cortez')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'NoahJ456', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('NoahJ456')
    ids.append(idcount)
    idcount += 1
    
#IT - 5 persons
twitobj = api.user_timeline(screen_name = 'elonmusk', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('elonmusk')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'The Onion', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('The Onion')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'The Babylon Bee', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('The Babylon Bee')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Andrej Karpathy', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Andrej Karpathy')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'PC Gamer', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('PC Gamer')
    ids.append(idcount)
    idcount += 1  
#random - 5 person
twitobj = api.user_timeline(screen_name = 'TTfue', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('TTfue')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'pokelawls', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('pokelawls')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'crutch', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('crutch')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Justin Sink', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Justin Sink')
    ids.append(idcount)
    idcount += 1
twitobj = api.user_timeline(screen_name = 'Eric Schultz', count = 20, include_rts = False)
for item in twitobj:
    tweets.append(item.text)
    usernames.append('Eric Schultz')
    ids.append(idcount)
    idcount += 1
    
# create pandas dataframe

ds = pd.DataFrame(list(zip(ids,usernames,tweets)), columns = ['id','Username','Tweet'])
print("Printing Dataframe")
print(ds)

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(ds['Tweet'])

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
print("Printing the cosine similarity matrix")
print("Matrix shape is: " + str(cosine_similarities.shape))
plt.imshow(cosine_similarities)
plt.colorbar()
plt.show()

results = {}

for idx, row in ds.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities[idx][i], ds['id'][i]) for i in similar_indices]

    results[row['id']] = similar_items[1:]
    
print('done!')

def item(id):
    return ds.loc[ds['id'] == id]['Tweet'].tolist()[0].split(' - ')[0]

# Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Recommending " + str(num) + " tweets similar to " + item(item_id) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: " + item(rec[1]) + " (score:" + str(rec[0]) + ")")

recommend(item_id=0, num=5)
# take some of the followers tweets to get more accuracy
```
